```python
import csv
from datetime import datetime
import matplotlib.pyplot as plt
filename = 'lincoln_weather.csv'
with open(filename) as f:
    reader = csv.reader(f)
    header_row = next(reader)
    print(header_row)
    prcps,dates = [],[]
    for row in reader:
        current_date = datetime.strptime(row[2],'%Y-%m-%d')
        prcp = row[5]
        dates.append(current_date)
        prcps.append(prcp)

        
plt.style.use('seaborn')
fig, ax = plt.subplots()
ax.plot(dates,prcps,c='red')

ax.set_title("Daily percpitation for lincoln ne, for 2022")
ax.set_xlabel('')
fig.autofmt_xdate()
ax.set_ylabel("percpitation (in)")
ax.tick_params(axis='both',which ='major',labelsize=10)
plt.show()
```
